```python
from collections import defaultdict
```

```python
parent = dict()
rank = dict()
def get_edges(g, V):
    edges = []
    for v in V:
        for to, cost in g[v]:
            edges.append((cost, v, to))
    edges = sorted(edges)
    return edges
def make_set(v):
    parent[v] = v
    rank[v] = 0

def find(v):
    if parent[v] != v:
        parent[v] = find(parent[v])
    return parent[v]

def union(v1, v2):
    root1 = find(v1)
    root2 = find(v2)
    if rank[root1] < rank[root2]:
        parent[root1] = root2
    elif rank[root1] > rank[root2]:
        parent[root2] = root1
    
    else:
        parent[root2] = root1
        rank[root1] += 1

def kruskal(graph, V):
    mst = defaultdict(set)
    for v in V:
        make_set(v)
        edges = get_edges(graph, V)
        edges.sort()
        #print edges
    for edge in edges:
        w, v1, v2 = edge
        if find(v1) != find(v2):
            union(v1, v2)
            mst[v1].add((v2, w))

    return dict(mst)

V = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
example_graph = {
    'A': {('B', 7), ('D', 5)},
    'B': {('A', 7), ('C', 8), ('D', 9), ('E', 7)},
    'C': {('B', 8), ('E', 5)},
    'D': {('A', 5), ('B', 9), ('E', 15), ('F', 6)},
    'E': {('B', 7), ('C', 5), ('D', 15), ('F', 8), ('G', 9)},
    'F': {('D', 6), ('E', 8), ('G', 11)},
    'G': {('E', 9), ('F', 11)},
}
```

```python
mst = kruskal(example_graph, V)
print(mst)
```
